1. Tentukan library digunakan

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

2. load dataset

```python
data = pd.read_csv('Customertravel.csv')
```

```python
data.head()
```

```python
data.isnull()
```

```python
data.shape
```

```python
data['Target'].value_counts()
```

```python
X = data.drop (columns='Target',axis=1)
Y = data['Target']
```

```python
print(X)
```

```python
print(Y)
```

3. Memisahkan data Training dan data Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

4. Membuat data latih menggunakan algoritma svm

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

5. membuat model evaluasi untuk mengukur tingkat akurasi

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Akurasi data training adalah =', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Akurasi data testing adalah=', test_data_accuracy)
```

6. Membuat model prediksi

```python
input_data = (30,1,2,1,1,1)

input_data_as_numpy_array = np.array(input_data)

input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

prediction = classifier.predict(input_data_reshape)
print(prediction)

if (prediction[0] == 0):
    print('Pelanggan Tidak Akan Bergeser')
else :
    print('Pelanggan Akan Bergeser')
```
